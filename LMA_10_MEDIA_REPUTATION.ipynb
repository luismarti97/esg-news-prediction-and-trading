```python
from rapidfuzz import process, fuzz
import numpy as np
```

```python
noticias = pd.read_csv('/content/drive/MyDrive/MIAX/TRABAJO FINAL DE MASTER DEFINITIVO/df_cleaned.csv')
```

```python
df_final = noticias.copy()
```

```python
sources = df_final["source"].dropna().unique().tolist()
grupos_fuzzy = {}
THRESHOLD = 90

for fuente in sources:
    match = process.extractOne(fuente, grupos_fuzzy.keys(), scorer=fuzz.token_sort_ratio)
    if match and match[1] > THRESHOLD:
        grupos_fuzzy[match[0]].append(fuente)
    else:
        grupos_fuzzy[fuente] = [fuente]

source_map = {}
for clave, grupo in grupos_fuzzy.items():
    nombre_base = min(grupo, key=len)
    for variante in grupo:
        source_map[variante] = nombre_base

df_final["sources"] = df_final["source"].map(source_map)
```

## Building a Reputation Score Based on Source Frequency

In this section, we introduce a mechanism to assign a **reputation score** to each media source within each sector, based on how often the source appears.

### Key operations:

- **Filtering only cited news**: We remove news articles with missing "sector" or "source" fields, as both are essential for reputation analysis.
- **Counting source appearances per sector**: Using `groupby(["sector", "source"]).size()`, we compute how many times each source is cited in each sector.
- **Sorting**: We sort the sources in descending order of frequency within each sector.

### Reputation score logic:

- A function `reputacion_escalada(x)` is created:
  - If a source has **30 or more** articles, it is assigned the **maximum reputation score (1)**.
  - If it has **fewer than 30** articles, its reputation score is **scaled between 0.5 and 1** using a **logarithmic scaling** based on the number of appearances.

The logarithmic scaling smooths the increase in reputation, ensuring that a small difference in article counts does not cause disproportionately large changes in reputation.

### Justification:

- **Reflects trustworthiness and relevance**: Highly cited sources are assumed to be more reputable or influential within their sector.
- **Avoids overfitting to extreme values**: The logarithmic transformation prevents large differences between medium and large sources.
- **Enables weighting in downstream models**: Later on, this reputation score can be used to modulate the impact of a news article depending on the reliability of its source.

This scoring system gives a nuanced and dynamically scaled reputation to media sources based on their empirical relevance, improving the robustness of subsequent analyses and models.

```python
def reputacion_escalada(x):
    if x >= 30:
        return 1
    else:
        return 0.5 + 0.5 * (np.log1p(x) / np.log1p(30))
top_sources_por_sector["reputation_score"] = top_sources_por_sector["count"].apply(reputacion_escalada)
```

```python
df_final = df_final.merge(
    top_sources_por_sector[["sector", "sources", "reputation_score"]],
    on=["sector", "sources"],
    how="left"
)
```
