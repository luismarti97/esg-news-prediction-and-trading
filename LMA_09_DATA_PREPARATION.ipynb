```python
df = pd.read_csv('/content/drive/MyDrive/MIAX/TRABAJO FINAL DE MASTER DEFINITIVO/df_finally_labeled.csv')
df
```

```python
columns_to_drop = ['sentiment_body', 'sentiment_body_score', 'description', 'esg_pred', 'id']
df_final = df.drop(columns=columns_to_drop)
df_final
```

```python
print(df_final['Publication Date'].dtype)
```

```python
df_final['Publication Date'] = df_final['Publication Date'].astype(str).str.slice(0, 10)
```

```python
df_final
```

```python
df_final['Publication Date'] = pd.to_datetime(df_final['Publication Date'], errors='coerce')
df_final
```

```python
df_final = df_final[df_final['ticker'] != 'ATVI'].reset_index(drop=True)
```

```python
df = df_final.dropna(subset=['title', 'content']).reset_index(drop=True)
```

```python
ticker_company_map = df.dropna(subset=['ticker', 'company']).drop_duplicates(subset=['ticker']).set_index('ticker')['company'].to_dict()
company_ticker_map = df.dropna(subset=['ticker', 'company']).drop_duplicates(subset=['company']).set_index('company')['ticker'].to_dict()

mask_company = df['company'].isna() & df['ticker'].notna()
df.loc[mask_company, 'company'] = df.loc[mask_company, 'ticker'].map(ticker_company_map)

mask_ticker = df['ticker'].isna() & df['company'].notna()
df.loc[mask_ticker, 'ticker'] = df.loc[mask_ticker, 'company'].map(company_ticker_map)
```

```python
df
```

```python
df['ticker'] = df['ticker'].replace('MRO', 'MPC')
```

```python
df[df['ticker'] == 'UNKNOWN']['company'].value_counts()
```

```python
corrections = {
    'disney': 'DIS',
    'general motors': 'GM',
    'jpmorgan': 'JPM',
    'morgan stanley': 'MS',
    'wells fargo': 'WFC',
    'comcast': 'CMCSA',
    'american express': 'AXP',
    'the home depot': 'HD'
}
mask_unknown = df['ticker'] == 'UNKNOWN'
df.loc[mask_unknown, 'ticker'] = df.loc[mask_unknown, 'company'].map(corrections).fillna('UNKNOWN')
```

```python
df['ticker'].value_counts()
```

```python
sector_mapping = df.dropna(subset=['sector']).groupby('ticker')['sector'].agg(lambda x: x.mode()[0]).to_dict()
mask = df['sector'].isna()
df.loc[mask, 'sector'] = df.loc[mask, 'ticker'].map(sector_mapping)
```

```python
df.isna().sum()
```

```python
columns_to_drop = ['company']
df = df.drop(columns=columns_to_drop)
df
```

```python
df['Publication Date'] = pd.to_datetime(df['Publication Date'])
df = df.sort_values(by='Publication Date').reset_index(drop=True)
df
```
